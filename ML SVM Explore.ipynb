```python
# import warnings
# warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

exo = pd.read_csv('exo_clean.csv')
exo.head()
```

```python
exo.columns
```

```python
target = exo['koi_disposition']
data = exo.drop(['koi_disposition'], axis=1)
feature_names = data.columns
data.head()
```

```python
# create train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Pre processing
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)
X_train_minmax = X_minmax.transform(X_train)
X_test_minmax = X_minmax.transform(X_test)
```

```python
print(X_minmax)
print(X_train_minmax)
print(X_test_minmax)
```

### Train the model (SVM)

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train_minmax, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_minmax, y_train)}")
print(f"Testing Data Score: {model.score(X_test_minmax, y_test)}")
```

### Hyperparameter tuning

```python
# gridsearchcv
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1,5,10,50],
              'gamma': [0.0001,0.0005,0.001,0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_minmax, y_train)
```

```python
# print best params and score
print(grid.best_params_)
print(grid.best_score_)
```

```python
# print training score
grid.score(X_train_minmax, y_train)
```

```python
# print testing score
grid.score(X_test_minmax,y_test)
```

```python
# make predictions
predictions = grid.predict(X_test_minmax)
print(predictions)
```

```python
# print classification
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
